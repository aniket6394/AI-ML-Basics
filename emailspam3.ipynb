```python
raw_email_data=pd.read_csv('/content/mail_data.csv')
```

```python
raw_email_data.shape
```

```python
#replace  the null values  with a null string
mail_data=raw_email_data.where((pd.notnull(raw_email_data)),'')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

```python
#label encoding
mail_data.loc[mail_data['Category']=='spam','Category',]=0
mail_data.loc[mail_data['Category']=='ham','Category',]=1
#spam-> 0 ham->1
```

```python
X=mail_data['Message']
Y=mail_data['Category']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42,test_size=0.2)
```

```python
#feature extraction
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
```

```python
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
```

```python
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
model=LogisticRegression()
model.fit(X_train_features,Y_train)
```

```python
# on training data
X_pred_train=model.predict(X_train_features)
print(classification_report(Y_train,X_pred_train))
```

```python
# on testing data
X_pred_test=model.predict(X_test_features)
print(classification_report(Y_test,X_pred_test))
print(confusion_matrix(Y_test,X_pred_test))
```
